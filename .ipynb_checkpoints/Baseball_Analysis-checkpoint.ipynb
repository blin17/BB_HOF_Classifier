```python
'''
@author: Brian Lin
'''
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.preprocessing import PolynomialFeatures

# Question 1: How has Average Salaries changed over time?
# Sum of Team Salaries / # of Teams
data_dir = 'baseballdatabank-master/core/'

# salaries ranges from 26 teams (1985-1992), 28 teams (93-97) to 30 (98-2015)
salary_df = pd.read_csv(data_dir + 'Salaries.csv')
team_df = pd.read_csv(data_dir + 'Teams.csv')
franchise_df = pd.read_csv(data_dir + 'TeamsFranchises.csv')
# adding in team and franchise information into salary df
salary_df = salary_df.merge(team_df[['yearID','teamID','franchID']], on=['teamID','yearID'], how= 'inner')
salary_df = salary_df.merge(franchise_df, on='franchID', how='inner')
```

```python
yearly_salary = salary_df.groupby(['yearID']).sum()
num_teams_in_a_year = salary_df.groupby(['yearID'])['teamID'].unique().apply(lambda d: float(len(d)))
avg_salary = (yearly_salary['salary']/ num_teams_in_a_year)
avg_salary_in_mil = avg_salary.apply(lambda d: round(d/ float(10**6),3))
avg_salary_in_mil.plot()
plt.title('Average Team Salarys Over Time')
plt.ylabel('Salary (in millions)')
plt.xlabel('Year')
```

```python
'''
Question 2: Can we make a logistic regression model using 
            number of awards and number of all star appearances to predict hall of fame chance?
'''
# num all star appearances by player
allstars_df = pd.read_csv(data_dir + 'AllstarFull.csv')[['playerID','yearID']].drop_duplicates()
allstars_df = allstars_df.groupby('playerID',as_index=False).count()

# removes all awards besides the "useful" ones
awards_df = pd.read_csv(data_dir + 'AwardsPlayers.csv')
useful_awards = (['Most Valuable Player', 'Cy Young Award', 'Rookie of the Year', 'All-Star Game MVP',
                  'World Series MVP', 'Silver Slugger', 'Gold Glove', 'Triple Crown','Pitching Triple Crown'])
useful_awards = (['Most Valuable Player', 'Cy Young Award', 'Rookie of the Year', 'All-Star Game MVP',
                  'World Series MVP', 'Silver Slugger', 'Gold Glove', 'Triple Crown','Pitching Triple Crown'])
awards_df = awards_df[awards_df.awardID.isin(useful_awards)]

# flattens the awards into an awards by player id
for award in useful_awards:
    award_name = '_'.join(map(lambda s: s.lower(), award.split()))
    awards_df[award_name] = (awards_df['awardID'] == award).astype(int)
awards_df = awards_df.groupby('playerID',as_index=False).sum().drop('yearID', axis='columns')
allstars_df = allstars_df.rename(index=str, columns={'yearID': 'all_stars'})

# merges both columns
data = awards_df.merge(allstars_df, on='playerID', how='outer')

# get all players and first year played and merges it to data
players_df = pd.read_csv(data_dir + 'Appearances.csv')[['playerID','yearID']].drop_duplicates()

# gets last year played of every player
players_df = players_df.groupby('playerID', as_index=False).min()
players_df = players_df.rename(columns={'yearID':'first_year_played'})

data = players_df.merge(data, on='playerID', how='outer')

# get hall of fame and merges it to data
hall_of_fame = pd.read_csv(data_dir + 'HallOfFame.csv')
hall_of_fame = hall_of_fame[(hall_of_fame['category'] == 'Player') & (hall_of_fame['inducted'] == 'Y')]
hall_of_fame['hall_of_fame'] = 1
hall_of_fame = hall_of_fame[['playerID','hall_of_fame']]
data = data.merge(hall_of_fame, on='playerID', how='left')
data = data.fillna(0)

# first year played of last player who joined in hall of fame, so we exclude everyone who joined the mlb after 1992
data[data['hall_of_fame'] == 1].sort_values(by = 'first_year_played', ascending= False)
data = data[data['first_year_played'] <= 1992]
#data = data.drop('first_year_played', axis = 'columns')
#data.sort_values(by = 'first_year_played', ascending=False)
```

```python
# removes the player names and first year played
training_data = data.iloc[:,2:].values
Xdata = training_data[:,:-1]
Ydata = training_data[:,-1]

# define cost function and gradient for logistic regression
def sigmoid(z):
    return np.power(1+np.exp(-z),-1)

def hypothesis(X, theta):
    return sigmoid(X.dot(theta))

def regularized_log_reg_cost(theta, X, Y, lamb = 1):
    m = float(len(X))
    theta_reg = lamb * (theta ** 2)
    theta_reg[0] = 0
    
    J = (-1/m  * (Y.dot(np.log(hypothesis(X,theta))) 
                  + (1-Y).dot(np.log(1-hypothesis(X,theta)))) + 1/(2 * m) * np.sum(theta_reg))
    
    if np.isnan(J):
        return(np.inf)
    return (J)

def regularized_log_reg_grad(theta, X, Y, lamb = 1):
    m = float(len(X))
    theta_reg = lamb * theta.copy()
    theta_reg[0] = 0
    
    grad = (X.T.dot(hypothesis(X,theta) - Y) + theta_reg)/ m
    return grad

# calculates accuracy, confusion matrix, and f1 score
def prediction_score(X, Y, theta,threshold=0.5):
    def greater_than_one_half(x):
        if x > threshold:
            return 1
        else:
            return 0
        
    predicted_X = [greater_than_one_half(x) for x in hypothesis(X, theta)]
    true_pos = 0
    false_pos = 0
    true_neg = 0
    false_neg = 0
    for i in range(len(predicted_X)):
        if predicted_X[i] == 1 and Y[i] == 1:
            true_pos +=1
        elif predicted_X[i] == 1 and Y[i] == 0:
            false_pos +=1
        elif predicted_X[i] == 0 and Y[i] == 1:
            true_neg +=1
        elif predicted_X[i] == 0 and Y[i] == 0:
            false_neg +=1
    
    wrong = float(np.sum(np.abs(predicted_X - Y)))
    total = float(len(X))
    ans = dict()
    ans['num_features'] = len(theta)
    ans['accuracy'] =  1-(wrong/total)
    ans['true_pos'] = true_pos
    ans['true_neg'] = true_neg
    ans['false_pos'] = false_pos
    ans['false_neg'] = false_neg
    if true_pos == 0 or false_pos == 0 or true_neg == 0:
        pass
    else:
        ans['precision'] = float(true_pos) / float((true_pos + false_pos))
        ans['recall'] = float(true_pos) / float((true_pos + true_neg))
        ans['f1_score'] = 2 * ans['precision'] * ans['recall'] / (ans['precision'] + ans['recall'] )
    return ans

# runs the function over the predefine range for polynomials, regulizarization values, and threshold values
accuracy_scores_df = []
for p in range(1,4):
    for reg in [0,1,100]:
        poly = PolynomialFeatures(p)
        Xdata_transformed = poly.fit_transform(Xdata)
        num_features = Xdata_transformed.shape[1]
        res2 = minimize(regularized_log_reg_cost,np.zeros((num_features,)), 
                        args=(Xdata_transformed,Ydata,reg), 
                        method=None, 
                        jac=regularized_log_reg_grad, 
                        options={'maxiter':30000})
        theta = res2.x
        for threshold_val in (np.divide(map(float,range(1,6)),10)):
            p_score = prediction_score(poly.fit_transform(Xdata), Ydata, theta,threshold_val)
            p_score['poly'] = p
            p_score['threshold'] = threshold_val
            p_score['reg'] = reg
            accuracy_scores_df.append(p_score)

# orders the columns and prints out the dataframe        
accuracy_scores_df = pd.DataFrame(accuracy_scores_df)      
accuracy_scores_df.iloc[accuracy_scores_df['f1_score'].argmax()]
cols = ['poly','threshold', 'reg', 'true_pos', 'false_pos','true_neg', 'false_neg','accuracy',  'precision', 'recall', 'f1_score']
accuracy_scores_df[cols]
```
